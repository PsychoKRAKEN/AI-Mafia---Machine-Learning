```python
g=pd.read_csv('Data\Assignment\Assignment 2\Testing Data\Test.csv')
```

```python
test_x=g.values
```

```python
from sklearn.linear_model import LinearRegression
```

```python
Lin=LinearRegression()
```

```python
Lin.fit(train_x,train_y)
```

```python
pred_y=Lin.predict(test_x)
```

```python
plt.plot(train_x,Lin.predict(train_x),color='blue',label='Prediction')
plt.plot(test_x,pred_y,color='black',label='Prediction')
plt.legend()
plt.show()
```

```python
df=pd.DataFrame(pred_y)
```
