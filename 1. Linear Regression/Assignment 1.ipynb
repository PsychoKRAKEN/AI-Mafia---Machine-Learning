```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fx=pd.read_csv('Data\Assignment\Assignment 1\Training Data\Linear_X_Train.csv')
fy=pd.read_csv('Data\Assignment\Assignment 1\Training Data\Linear_Y_Train.csv')
```

```python
train_x=fx.values
train_y=fy.values
```

```python
train_x=train_x.reshape((-1,1))
train_y=train_y.reshape((-1,1))
```

```python
plt.scatter(train_x,train_y)
plt.show()
```

```python
gx=pd.read_csv('Data\Assignment\Assignment 1\Testing Data\Linear_X_Test.csv')
```

```python
test_x=gx.values.reshape((-1,1))
```

```python
from sklearn.linear_model import LinearRegression
```

```python
Lin=LinearRegression()
```

```python
Lin.fit(train_x,train_y)
```

```python
pred_y=Lin.predict(test_x)
```

```python
plt.scatter(train_x,train_y,color='red',label='Data')
plt.plot(train_x,Lin.predict(train_x),color='black',label='Prediction')
plt.legend()
plt.show()
```

```python
plt.plot(test_x,pred_y,color='black',label='Prediction')
plt.show()
```

```python
df=pd.DataFrame(pred_y)
```

```python
df.to_csv('Data\Assignment\Assignment 1\Predict_y.csv',index=False)
```
